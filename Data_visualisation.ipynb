**Subplots**

Subplot -- used to display more then one graph side-by-side, the object who contains all the subplots is called a figure, we can have many subplots in one figure. The command used to create subplots **subplot()** and needs three arguments to be passed into it:
- The number of rows of subplots
- The number of columns of subplots
- The index of the subplot we want to create

Any plt.plot() that comes after plt.subplot() will create a line plot in the specified subplot. For instance:

```python
x = [1, 2, 3, 4]
y = [1, 2, 3, 4]
#first subplot
plt.subplot(1, 2, 1)
plt.plot(x, y, color = 'green')
plt.title('First subplot')
# Second subplot
plt.subplot(1, 2, 2)
plt.plot(x, y, color='steelblue')
plt.title('Second Subplot')

plt.show()
```

**Subplots Part II**

We can customize the spacing between our subplots to make sure that the figure we create is visible and easy to understand.

**plt.subplots_adjust()** command has some keywords arguments:

* left  the left-side margin, with a default of 0.125. You can increase this number to make room for a y-axis label
* right  the right-side margin, with a default of 0.9. You can increase this to make more room for the figure, or decrease it to make room for a legend
* bottom  the bottom margin, with a default of 0.1. You can increase this to make room for tick mark labels or an x-axis label
* top  the top margin, with a default of 0.9
* wspace  the horizontal space between adjacent subplots, with a default of 0.2
* hspace  the vertical space between adjacent subplots, with a default of 0.2

```python
# left plot
plt.subplot(1, 2, 1)
plt.plot([-2, -1, 0, 1, 2], [4, 1, 0, 1, 4])
plt.title('First subplot')
plt.ylabel('practice y axis')

#right plot 
plt.subplot(1,2,2)
plt.plot([-2, -1, 0, 1, 2], [4, 1, 0, 1, 4])
plt.title('Second subplot')
plt.ylabel('practice y axis')
# subplot adjust
plt.subplots_adjust(bottom = 0.4, top = 1, wspace = 0.5,  left = 0.2)
#plt.subplots_adjust(top=0.95, hspace=0.25, wspace = 0.5)
plt.show()
```

**Ticks**

Due to the fact that plot can have many subplots, we have to specify which one we want to modify, in order to do that we call subplot() in a different way:

- then specify which one we want to modify  **ax.set_xticks([  ])**
- ax is an axes oject and **lets us modify the axes belonging to a specific subplot**
- We can also modify the y-ticks by using ax.set_yticks().
- But, if we want special labels (such as strings), we can use the command **ax.set_xticklabels() or ax.set_yticklabels()**.

For example we might want to have y-axis with ticks at 10, 30, 50, 60 but label them '10%', '20%', '30%', and '40%' respectively. To do this we do following command. 

```python
ax = plt.subplot(1, 1, 1)
plt.plot([0, 1, 2, 3, 4], [0, 1, 4, 9, 16], marker = 'o' )
plt.plot([0, 1, 2, 3, 4], [0, 1, 8, 27, 64], marker = 's')
ax.set_xticks([i * 0.5 for i in range(1, 9)])
ax.set_yticks([10, 30, 50, 60])
ax.set_yticklabels(['10%', '20%', '30%', '40%'])#, '60%', '70%', '80%', '90%'])
plt.legend(['y-axis labeling'])
plt.show()
```

**Side-by-side chart bars**

We can use a bar chart to compare two sets of data with the **same types of axis values**. To do this, we plot two sets of bars next to each other, so that the values of each category can be compared. For example,
- comparing the population between two countries, 
- prices for two different foods , 
- enrollments in different claseses for males and females

To arrange bars side by side requires some math, so generate two list with this formula by make use of list comprehension

```python
#by nextcode compare different location of the MatplotSlip coffee empire
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales1 =  [91, 76, 56, 66, 52, 27]
sales2 = [65, 82, 36, 68, 38, 40]

n1 = 1  # This is our first dataset (out of 2)
n2 = 2  # This is our second dataset (out of 2)
t = 2 # Number of datasets
d = 6 # Number of sets of bars
w = 0.8 # Width of each bar
x1 = [t*element + w*n1 for element in range(d)]
x2 = [t*element + w*n2 for element in range(d)]
middle_x = [(a + b) / 2 for a, b in zip(x1, x2)]
plt.bar(x1, sales1)
plt.bar(x2, sales2)
plt.legend(['sale1', 'sale2'])
plt.xticks(middle_x, drinks, rotation = 30)
plt.show()
```

**Error Bars** --> Display error on bars

In the previous exercise, you learned to represent data as bars of different heights. Sometimes, we need to visually communicate some sort of **uncertainty in the heights of those bars**.

Here are some examples:
- The average number of students in a 3rd grade classroom is 30, but some classes have as few as 18 and others have as many as 35 students.
- We measured that the weight of a certain fruit was 35g, but we know that our scale isnt very precise, so the true weight of the fruit might be as much as 40g or as little as 30g.
- The average price of a soda is 1.00, but we also want to communicate that the standard deviation is 0.20.

**plt.bar(x, y, yerr = 2, capsize = 10)**: If we wanted to show an error of +/- 2, we would add the keyword yerr=2 to our plt.bar command. To make the caps wide and easy to read, we would add the keyword capsize=10. If we want a different amount of error for each bar, we can make **yerr** equal to a list rather than a single number:

```python
values = [10, 13, 11, 15, 20]
yerr = [1, 3, 0.5, 2, 4]
x = range(len(values))
plt.figure(figsize = (7, 4))
plt.bar(x, values, yerr = yerr, capsize = 10, width = 0.5)
plt.show()
```

```python
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
ounces_of_milk = [6, 9, 4, 0, 9, 0]
error = [0.6, 0.9, 0.4, 0, 0.9, 0]
x = range(len(drinks))

plt.figure(figsize = (5, 4))
plt.bar(x, ounces_of_milk, yerr = error, capsize = 10)
plt.show()
```

**Fill Between**

Matplotlib, we can use **plt.fill_between** to shade error. This function takes three arguments:

1. x-values  this works just like the x-values of plt.plot
2. lower-bound for y-values  sets the bottom of the shared area
3. upper-bound for y-values  sets the top of the shared area



```python
x_values = range(10)
y_values = [10, 12, 13, 13, 15, 19, 20, 22, 23, 29]
y_lower = [8, 10, 11, 11, 13, 17, 18, 20, 21, 27]
y_upper = [12, 14, 15, 15, 17, 21, 22, 24, 25, 31]

plt.fill_between(x_values, y_lower, y_upper, alpha = 0.2)
plt.plot(x_values, y_values)
plt.show()
```

**Pie Chart**

Pie charts = useful when we want to **display elements of a data set as proportion of a whole**.

Pie charts are helpful for displaying data like:

- Different ethnicities that make up a school district
- Different macronutrients (carbohydrates, fat, protein) that make up a meal
- Different responses to an online poll

Option for pie charts:
- labeling the each category, understand what each sloce represents 
  1. use a legend to label each color, or
  2. put labels on the chart itself.
- another option of labeling is by adding the percentage of the total that each slice ocupies, this can be done automatically by Matplotlib with keyword **autopct** and we pass **string formatting** instructions to format the labels how we want. 

- '%0.2f'  2 decimal places, like 4.08
- '%0.2f%%'  2 decimal places, but with a percent sign at the end, like 4.08%. You need two consecutive percent signs because the first one acts as an escape character, so that the second one gets displayed on the chart.
- '%d%%'  rounded to the nearest int and with a percent sign at the end, like 4%.


```python
# method 1 for labeling and displaying pie charts 
budget_data = [500, 1000, 750, 300, 100]
budget_categories = ['marketing', 'payroll', 'engineering', 'design', 'misc']

plt.pie(budget_data, autopct = '%0.1f%%')
plt.axis('equal')
plt.legend(budget_categories)
plt.show()
```

```python
# method 2 for labeling and displaying pie charts
budget_data = [500, 1000, 750, 300, 100]
budget_categories = ['marketing', 'payroll', 'engineering', 'design', 'misc']

plt.pie(budget_data, labels = budget_categories, autopct = '%d%%')
plt.axis('equal')
plt.show()
```

**Histograms**

**Histograms are best for showing the shape of a dataset. For example, you might see that values are close together, or skewed to one side**

A histograms tells you **how many values in a datasets fall to different sets of numbers** (i.e how many numbers fall between 0 and 10? between 10 and 20? between 20 and 30?)
- all **bins** in a histogram are always the same size, the width of each bin is the distance between the minimum and maximum values of each bin. If we want to customize the number a bins in a dataset we can use keyword **bins** to set how many bins we want to divide the data into.
- each bin is represented by a different rectangle, whose height is the numeber of elements from the dataset fall within that bin 
- keyword **range** selects the minimum and maximum values to plot, for example if we want to display just a portion from a dataset we can displayed this values with  keyword **range = (number_1, number_2)**

**Multiple Histograms**

We can plot multiple histograms omn the same plot, this could be useful for example, comparing the heights of a group of male and the heights of a bunch of womwn. 

- keyword **alpha** = **0 or 1**, which can be value between 0 and 1, sets the transparency of the histogram. 0 = transparent going to 1 get opaque
- keyword **histtype** = **'step'** to draw just the outline of a histogram 
- anoother problem we face is that our histogram might have different numbers of samples, making one much bigger than the another. To solve this, we can normalize our histograms using **normed = True**, This comands divides the height of each column by a constant such that the total shaded area of the histograms sums 1 

```python
x1 = random.normal(loc=1, scale=1.5, size= 1000)
x2 = random.normal(loc=1, scale=2, size= 500)
```

```python
plt.hist(x1, bins = 20, alpha = 0.8, color = 'green', edgecolor = 'red', histtype = 'step', range = (-6, 2))
plt.hist(x2, bins = 20, alpha = 0.8, edgecolor = 'black', histtype = 'step')
plt.show()
```

```python
plt.hist(x1, bins = 20, alpha = 0.5, color = 'green', edgecolor = 'red', density = True)
plt.hist(x2, bins = 20, alpha = 0.5, edgecolor = 'black', density = True)
plt.show()
```

**Seaborn multiple distribution**

In this lesson, we will explore how to use Seaborn to graph multiple statistical distributions, including box plots and violin plots.

**KDE Plots, Part I**

How can we get more insights from dataset?
- data could be clustered arround the mean or spred out evenly accross the entire range 
- to get more information about each of these datasets well need to examine their distribution 
- one way is to use histograms, but they have their drawbacks, too 

Seaborn comes with graphoing distribution **KDE (Kernel Density Estimator)**

- What are the basic of KDE?

It is a **probability density function**, that adds centered kernel points an eventually ads all together 

**Advantage of KDE**
- the smooth the dataset
- allows to generilize over the shape of our data 
- aren't beholden to specific datapoints

KDE plots are preferable to histograms because depending on how you group the data into bins and the width of the bins, you can draw widely different conclusions about the shape of the data. 

How to use it?
- data - the univariate dataset being visualized, like a Pandas DataFrame, Python list, or NumPy array
- shade - a boolean that determines whether or not the space underneath the curve is shaded

```python
dataset2 = df['age']
dataset3 = df['hours-per-week']
sns.set_style("darkgrid")
sns.set_palette("pastel")
sns.kdeplot(dataset2, shade=True, label="dataset2")
sns.kdeplot(dataset3, shade=True, label="dataset3")
plt.legend()
plt.show()
```

**Box Plots**

While KDE can tell us about the shape of the data, it is cumbersome  to comapare multiple KDE plots at once.

What is a box plot?
- the box plots can tell us about how our dataset is distributed like KDE plot, 
- but it shows the range of our dataset, 
- where a significant portion of data lies
- and wheather or not any outliers are present 

Lets examine how we interpret a box plot:

- The box represents the interquartile range
- The line in the middle of the box is the median
- The end lines are the first and third quartiles
- The diamonds show outliers

interquartile range = called midspread, middle 50%, is measure of statistical dispersion, being equal to the difference between 75th and 25th 

**Advantage**, one advantage of the box plot over the KDE plot is that in seaborn it is easy to plot multiple and compare distributions.

**Disadvantage** do not display bimodal dataset

What are the basic boxplots arguments?
- data - the dataset were plotting, like a DataFrame, list, or an array
- x - a one-dimensional set of values, like a Series, list, or array
- y - a second set of one-dimensional data



```python
sns.boxplot(data = df, x = 'race', y = 'age')
plt.xticks(rotation = 45)
plt.show()
```

**Violin plots**

What is a violin plot?
 - A violin plot combines the ability to visualize the shape of the distribution, along with showing statistical information like the median.
- comparing distribution 
- it is a combination of KDE and box plots 

**advantages**
- provides more information than box plots because instead of maping each individual data point, we get an estimation of the data set thanks to KDE

**disadvantage**
- are tricker to read

How to interpret violin plots?

- There are two **KDE plots** that are symmetrical along the center line.
- A **white dot** represents the median.
- The **thick black line** in the center of each violin represents the interquartile range.
- The **lines that extend from the center** are the **confidence intervals** - just as we saw on the bar plots, a violin plot also displays the 95% confidence interval.

What information to passing in sns.violinplot()?
- data - the dataset that were plotting, such as a list, DataFrame, or array
- x, y, and hue - a one-dimensional set of data, such as a Series, list, or array
- any of the parameters to the function sns.boxplot()


```python
sns.violinplot(data = df, x = 'race', y = 'age')
plt.xticks(rotation = 45)
plt.show()
```

**Review**
- KDE plot - Kernel density estimator; shows a smoothed version of dataset. Use sns.kdeplot().
- Box plot - A classic statistical model that shows the median, interquartile range, and outliers. Use sns.boxplot().
- Violin plot - A combination of a KDE and a box plot. Good for showing multiple distributions at a time. Use sns.violinplot().
