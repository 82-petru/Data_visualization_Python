# Visualizing Kiva Data With Seaborn

Dataset belongs from kaggle, kaggle allows user to publish and find data-set, this web-based has a data-science profile 

Data-set contains information about loans awarded by the non-profit Kiva. Kiva este o organizatie prin care faci un imprumut de o suma mica, functioneaza pe principiul crowdfunding care este o practica de a finanta un proiect prin solicitarea de fonduri mici de la un numar mare de oameni.

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.ticker as mtick
```

```python
df = pd.read_csv('kiva_data.csv')
df.head()
```

**Explore the average loan amounts within countries by gender using aggregated bar charts**

```python
f, ax = plt.subplots(figsize = (10, 5))
sns.barplot(data = df, x = 'country', y = 'loan_amount', hue = 'gender')
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
```

**Explore the number of loan by countries using aggregates bar charts** 

```python
sns.barplot(data = df, x = 'activity', y = 'loan_amount', estimator = len)
plt.show()
```

**Visualize the distribution of loan amount by project**

```python
# visualize by kde plots 
f, ax = plt.subplots(figsize = (10, 6))
female = df[df.gender == 'female']
male = df[df.gender == 'male']
sns.set_style('white')
sns.set_palette('Accent')
sns.kdeplot(female['loan_amount'], shade = True, label = 'female')
sns.kdeplot(male['loan_amount'], shade = True, label = 'male')
ax.set_title('Loan amount by project')
ax.set_xlabel('Loans number')
plt.show()
```

```python
# visualize by boxplots 
plt.figure(figsize = (4, 6))
sns.set_style('darkgrid')
sns.boxplot(data = df, x = 'gender', y = 'loan_amount')
plt.show()
```

```python
#visualize by violin
plt.figure(figsize = (8, 6))
sns.set_style('ticks')
sns.set_palette('Set2')
sns.set_context('poster', font_scale = 0.6, rc = {'grid.linewidith' : 2})
sns.violinplot(data = df, x = 'gender', y = 'loan_amount')
sns.despine(left = True, bottom = True)
plt.show()
```

**Visualize distributions of loans by country using boxplots**


```python
f, ax = plt.subplots(figsize = (12, 8))
sns.boxplot(data = df, x = 'country', y = 'loan_amount')
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
plt.show()
```

**Visualize loans by activity  using boxplots**

```python
f, ax = plt.subplots(figsize = (10, 7))
sns.set_palette('Dark2')
sns.boxplot(data = df, x = 'activity', y = 'loan_amount')
plt.show()
```

**Visualize distributions of loans within countries by gender using violin**

```python
f, ax = plt.subplots(figsize = (13, 7))
sns.set_palette('Spectral')
sns.violinplot(data = df, x = 'country', y = 'loan_amount', hue = 'gender', split = True)
plt.show()
```
