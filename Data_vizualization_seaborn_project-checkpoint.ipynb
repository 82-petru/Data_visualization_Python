```python
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt
import seaborn as sns
```

1. **Load the csv file and create dataframe called df and inspect data using head()**

```python
df = pd.read_csv('WorldCupMatches.csv')
df.head(15)
```

2. **create a new column with name 'Total Goals' and set it equal to the sum of the columns home team goals and away team goals, display the results to confirm the column.**

```python
df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']
df.head()
```

3. **create a bar chart visualizing how many goals were scored each year the world cup was held between 1930 - 2014?**

```python
f, ax = plt.subplots(figsize = (12, 7))
sns.set_style('whitegrid')
ax = sns.barplot(data = df, x = 'Year', y = 'Total Goals', ci = None, estimator = np.sum)
sns.set_context('poster', font_scale = 0.5)
plt.xticks(rotation = 60)
sns.despine(bottom = True, left = True)
plt.show()
plt.close('all')
```

```python
home = df[['Home Team Name', 'Home Team Goals']]
draft = df[['Away Team Name', 'Away Team Goals']]
away = draft.rename(columns = {'Away Team Name' : 'Home Team Name', 'Away Team Goals' : 'Home Team Goals' })
frames = [home, away]
all_countries = pd.concat(frames).reset_index(drop = True)
all_countries.head()
```

```python
countries = all_countries.rename(columns = {'Home Team Name' : 'Country', 'Home Team Goals' : 'Goals'})
countries.head()
```

```python
ax = plt.subplots(figsize = (16, 7))
sns.set_style('whitegrid')
sns.set_context('paper', font_scale = 1.5)
ax = sns.barplot(data = countries, x = 'Country', y = 'Goals', ci = None, estimator = np.sum)
plt.xticks(rotation = 90)
plt.show()
```

```python
total_goals = countries.groupby('Country').Goals.count().reset_index()
total_goals[total_goals.Country == 'Romania']
```
